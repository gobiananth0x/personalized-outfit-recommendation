```python
from ultralytics import YOLO
```

```python
model = YOLO("yolo26n-cls.pt")
```

```python
results = model.train(
    data="/kaggle/input/dress-dataset/dataset",
    epochs=60,
    imgsz=224,
    batch=32,
    optimizer="AdamW",
    lr0=0.0005,
    patience=10,
    device=0,
    workers=2,
    cache=True,
    amp=True,
)
```

```python
model = YOLO("/kaggle/working/runs/classify/train/weights/best.pt")
```

```python
img = "/kaggle/input/dress-dataset/dataset/val/T-Shirt/06de9ebd-be39-4cc6-aea7-1d047ccf1a52.jpg"

result = model.predict(source=img, imgsz=224, conf=0.90)

print(f"Result: {model.names[int(result[0].probs.top1)]}")
```
